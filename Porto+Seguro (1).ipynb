Get dummies for the categorical variables, and split the data into positive and negative parts.

```python
train=train.drop(['id'],axis=1)
dummies=['ps_ind_02_cat','ps_ind_04_cat','ps_ind_05_cat','ps_car_01_cat','ps_car_02_cat','ps_car_03_cat','ps_car_04_cat',
         'ps_car_05_cat','ps_car_06_cat','ps_car_07_cat','ps_car_09_cat','ps_car_10_cat','ps_car_11_cat']
train_with_dummies= pd.get_dummies(train,  columns=dummies)
X_0=train_with_dummies[train_with_dummies['target']==0]
X_1=train_with_dummies[train_with_dummies['target']==1]
X_1.head()
```

Perform t-test on the same columns in X_1 and X_0. Keep only columns with p<0.05

We can use this test, if we observe two independent samples from the same or different population, e.g. exam scores of boys and girls or of two ethnic groups. The test measures whether the average (expected) value differs significantly across samples. If we observe a large p-value, for example larger than 0.05 or 0.1, then we cannot reject the null hypothesis of identical average scores. If the p-value is smaller than the threshold, e.g. 1%, 5% or 10%, then we reject the null hypothesis of equal averages.

```python
columns=X_1.columns
selected_columns=[]
for n in range(len(columns)):
    print('p-value:')
    print(columns[n])
    print(stats.ttest_ind(X_0[columns[n]], X_1[columns[n]], equal_var = False)[1])
    print(" ")
    if not columns[n]=='target':
        if stats.ttest_ind(X_0[columns[n]], X_1[columns[n]], equal_var = False)[1]<0.05:
            selected_columns.append(columns[n])

selected_columns
```

Undersampling the X_0 part. The positive obs is about 25000. I have tried different values for X=0 from 25000 to 55000, and 45000 is the optimal value.

```python
X_0=X_0.sample(n=45000)
X_0=X_0.append(X_1)
X=X_0[selected_columns]
Y=X_0['target']
```

```python
x_for_prediction=test.drop(['id'],axis=1)
x_for_prediction_with_dummies= pd.get_dummies(x_for_prediction,  columns=dummies)
x_for_prediction_with_dummies=x_for_prediction_with_dummies[selected_columns]
x_for_prediction_with_dummies.head()
```

```python
#ML part
#trainning set: X, Y; test set: x_for_prediction_with_dummies
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.5, stratify=Y, random_state=0)
print(len(X_train))
```

```python
from sklearn.linear_model import LinearRegression
#normalize is an improvement from 0.249 to 0.251 on LB
model = LinearRegression(normalize=True)
model.fit(X_train, y_train)
X_predicted=model.predict(X_test)

for i in range(len(X_predicted)):
    if X_predicted[i]>1:
        X_predicted[i]=1
    if X_predicted[i]<0:
        X_predicted[i]=0

print("gini_normalized")
print(gini_normalized(y_test, X_predicted))
print('gini')
print(gini(y_test, X_predicted))
```

```python
import sys
import numpy as np
import pandas as pd
def show_mem_usage():
    '''Displays memory usage from inspection
    of global variables in this notebook'''
    gl = sys._getframe(1).f_globals
    vars= {}
    for k,v in list(gl.items()):
        # for pandas dataframes
        if hasattr(v, 'memory_usage'):
            mem = v.memory_usage(deep=True)
            if not np.isscalar(mem):
                mem = mem.sum()
            vars.setdefault(id(v),[mem]).append(k)
        # work around for a bug
        elif isinstance(v,pd.Panel):
            v = v.values
        vars.setdefault(id(v),[sys.getsizeof(v)]).append(k)
    total = 0
    for k,(value,*names) in vars.items():
        if value>1e6:
            print(names,"%.3fMB"%(value/1e6))
        total += value
    print("%.3fMB"%(total/1e6))

show_mem_usage()


#import sys
#if sys.maxsize > 2**32:
#    print('64-bit')
#else:
#    print('32-bit')
```
